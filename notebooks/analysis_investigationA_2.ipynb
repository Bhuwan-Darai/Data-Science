```python
# 2_analysis_investigationA.ipynb
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Perform statistical analysis for Investigation A (bat vs rat behavior).
# Load cleaned data
bat_df = pd.read_csv("../data/bat_cleaned.csv")

# Investigation A  Do bats perceive rats as predators?

# Boxplot: risk vs seconds after rat arrival
plt.figure(figsize=(8,5))
sns.boxplot(data=bat_df, x='risk', y='seconds_after_rat_arrival')
plt.title("Seconds After Rat Arrival vs Bat Risk-Taking")
plt.xlabel("Risk (0 = Avoidance, 1 = Taking Risk)")
plt.ylabel("Seconds After Rat Arrival")
plt.savefig("../report/Figures/risk_vs_seconds_after_rat.png", dpi=300, bbox_inches='tight') # Save the image 
plt.show()

# Correlation between risk and reward
corr, p = pearsonr(bat_df['risk'], bat_df['reward'])
print(f"Correlation between risk and reward = {corr:.2f}, p-value = {p:.4f}")

# Scatter: rat presence duration vs risk
plt.figure(figsize=(8,5))
sns.scatterplot(data=bat_df, x='rat_duration', y='risk', alpha=0.6)
plt.title("Rat Presence Duration vs Bat Risk")
plt.xlabel("Rat Duration (seconds)")
plt.ylabel("Risk (0 = Avoidance, 1 = Taking Risk)")
plt.savefig("../report/Figures/rat_duration_vs_risk.png", dpi=300, bbox_inches='tight')
plt.show()
```
