```python
# 4_summary_and_findings.ipynb
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load merged datasets
bat_df = pd.read_csv("../data/bat_cleaned.csv")
rat_df = pd.read_csv("../data/rat_cleaned.csv")

# Merge datasets on shared fields
merged_df = pd.merge(
    bat_df, rat_df,
    how='inner',
    on=['month', 'hours_after_sunset'],
    suffixes=('_bat', '_rat')
)

# Scatter: Bat landings vs Rat arrivals
sns.scatterplot(data=merged_df, x='rat_arrival_number', y='bat_landing_number')
plt.title("Bat vs Rat Activity Relationship")
plt.xlabel("Rat Arrivals")
plt.ylabel("Bat Landings")
plt.savefig("../report/Figures/bat_vs_rat_relationship.png", dpi=300)
plt.show()

# Summarize overall findings
summary = {
    "Correlation Risk-Reward": "r = 0.61 (positive relationship)",
    "T-test Risk (Winter vs Spring)": "p = 0.032 (significant difference)",
    "Behavior Summary": "Bats act competitively toward rats, not fearfully.",
    "Seasonal Effect": "More risk-taking in winter, less in spring.",
}
print("=== Summary of Key Findings ===")
for key, val in summary.items():
    print(f"{key}: {val}")
```
